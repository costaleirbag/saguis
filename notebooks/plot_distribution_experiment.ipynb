```python
import requests
from PIL import Image
from io import BytesIO

import matplotlib.pyplot as plt

# Fetch the 10 image URLs
image_urls = df[df['scientific_name'] == 'Callithrix']['image_url'].dropna().sample(10)

# Create a figure
fig, axes = plt.subplots(2, 5, figsize=(15, 6))
axes = axes.flatten()

# Loop through the URLs and plot each image
for ax, url in zip(axes, image_urls):
    try:
        response = requests.get(url)
        img = Image.open(BytesIO(response.content))
        ax.imshow(img)
        ax.axis('off')
    except Exception as e:
        ax.text(0.5, 0.5, 'Image not available', ha='center', va='center')
        ax.axis('off')

plt.tight_layout()
plt.show()
```

```python
df_sudeste = df[df['place_state_name'].isin(['São Paulo', 'Rio de Janeiro', 'Minas Gerais', 'Espírito Santo'])]
```

```python
df_agrupado = df.groupby(['place_state_name', 'scientific_name']).size().reset_index(name='counts')
```

```python
df_sudeste_agrupado = df_sudeste.groupby(['place_state_name', 'scientific_name']).size().reset_index(name='counts').sort_values(by='counts', ascending=False)
```

```python
# Bar plot of df_sudeste_agrupado
import seaborn as sns
plt.figure(figsize=(12, 6), dpi=250)
sns.barplot(data=df_sudeste_agrupado, x='place_state_name', y='counts', hue='scientific_name')
plt.title('Observações de Callithrix por Estado no Sudeste')
plt.xlabel('Estado')
plt.ylabel('Número de Observações')
plt.legend(title='Espécie (scientific_name)', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45)
plt.tight_layout()


plt.savefig('../data/figures/observacoes_callithrix_sudeste.png', dpi=250)
plt.show()
```

```python
df_sudeste_agrupado_filtered = df_sudeste_agrupado[~df_sudeste_agrupado['scientific_name'].isin(['Callithrix geoffroyi', 'Callithrix flaviceps', 'Callithrix kuhlii'])]

plt.figure(figsize=(12, 6), dpi=250)
sns.barplot(data=df_sudeste_agrupado_filtered, x='place_state_name', y='counts', hue='scientific_name')
plt.title('Observações de Callithrix por Estado no Sudeste')
plt.xlabel('Estado')
plt.ylabel('Número de Observações')
plt.legend(title='Espécie (scientific_name)', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45)
plt.tight_layout()


plt.savefig('../data/figures/observacoes_callithrix_sudeste_filtered.png', dpi=250)
plt.show()
```

```python
df_sudeste_filtered = df_sudeste.copy()
df_sudeste_filtered['scientific_name'] = df_sudeste['scientific_name'].replace(['Callithrix geoffroyi', 'Callithrix flaviceps', 'Callithrix kuhlii'], 'Outras')
df_sudeste_agrupado_filtered = df_sudeste_filtered.groupby(['place_state_name', 'scientific_name']).size().reset_index(name='counts').sort_values(by='counts', ascending=False)
```

```python
import matplotlib.pyplot as plt
from matplotlib.patches import Patch
import matplotlib.cm as cm

# Use a large color palette (e.g., 'tab20' repeated or 'nipy_spectral')
species = df_sudeste_agrupado_filtered['scientific_name'].unique()
n_species = len(species)

# Use a continuous colormap (more than 20 colors) and sample from it
colormap = cm.get_cmap('Set2', n_species)
color_map = {name: colormap(i) for i, name in enumerate(species)}

# Setup subplots
unique_states = ['São Paulo', 'Rio de Janeiro', 'Minas Gerais', 'Espírito Santo']
fig, axes = plt.subplots(2, len(unique_states)//2, figsize=(18, 16), dpi=150, subplot_kw={'aspect': 'equal'})

# Helper to suppress small labels
def autopct_generator(threshold=5):
    def inner_autopct(pct):
        return f'{pct:.1f}%' if pct > threshold else ''
    return inner_autopct

# Flatten the axes array to iterate over individual Axes objects
axes = axes.flatten()

for ax, (state, group) in zip(axes, df_sudeste_agrupado_filtered.groupby('place_state_name')):
    group = group.set_index('scientific_name')
    ax.pie(
        group['counts'],
        labels=None,
        autopct=autopct_generator(threshold=5),  # Only show % if > 5%
        startangle=90,
        colors=[color_map[name] for name in group.index],
        # Change size of pctext
        textprops={'fontsize': 16, 'color': 'black'},
    )
    ax.set_title(f'{state}', fontsize=19)
    ax.set_ylabel('')
    ax.set_xticks([])
    ax.set_yticks([])

# Create legend handles and labels separately
legend_elements = [Patch(facecolor=color_map[name]) for name in species]
legend_labels = [name for name in species]

fig.legend(legend_elements, legend_labels, loc='upper center', bbox_to_anchor=(0.5, 1.08),
           ncol=4, frameon=False, title='Scientific Name', fontsize=19, title_fontsize=26)

plt.tight_layout()
plt.show()
```

```python
# Save it as png without background
fig.savefig('../data/figures/observacoes_callithrix_sudeste_pie.png', dpi=150, bbox_inches='tight', transparent=True)
# Save it as svg
fig.savefig('../data/figures/observacoes_callithrix_sudeste_pie.svg', dpi=150, bbox_inches='tight', transparent=True)
```
