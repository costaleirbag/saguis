```python
import pandas as pd

specie = 'jacchus'

path = f'../data/csv/raw_{specie}.csv'

df = pd.read_csv(path)
df
```

```python
df = df.dropna()
```

```python
df.columns = [x.lower() for x in df.columns]
```

```python
if df.columns.tolist() != ['url', 'classe']:
    df = df[['image_url', 'tipo']].rename(columns={'image_url': 'url', 'tipo': 'classe'})

df
```

```python
df.to_csv(path, index=False)
```
